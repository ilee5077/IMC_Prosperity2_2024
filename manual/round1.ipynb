### problem statement

With a large school of goldfish visiting, an opportunity arises to acquire some top grade SCUBA_GEAR. You only have two chances to offer a good price. Each one of the goldfish will accept the lowest bid that is over their reserve price. You know there’s a constant desire for scuba gear on the archipelago. So, at the end of the round, you’ll be able to sell them for 1000 SeaShells a piece.

Whilst not every goldfish has the same reserve price, you know the distribution of their reserve prices. The reserve price will be no lower than 900 and no higher than 1000. The probability scales linearly from 0 at 900 to most likely at 1000.

You only trade with the goldfish. Bids of other participants will not affect your results.

Think hard about how you want to set your two bids in order to walk away with some serious SeaShell profit.

```python
import numpy as np
import matplotlib.pyplot as plt

h = plt.hist(np.random.triangular(900, 1000, 1000, 100000), bins=200,
             density=True)
plt.show()
```

```python
best_offer1_list = []
best_offer2_list = []
max_profit_list = []
for i in range(1,5000):
    np.random.seed(i)
    reserve_prices = np.random.triangular(900, 1000, 1000, size = 5000)
    max_profit = 0
    #from running few iterations I knew offer1 was around 950, so reducing the search size
    for offer1 in range(940,960): 
        for offer2 in range(offer1,1001):
            profit = 0
            for reserve in reserve_prices:
                if offer1 >= reserve:
                    profit += 1000-offer1
                elif offer2 >= reserve:
                    profit += 1000-offer2

            if profit > max_profit:
                max_profit= profit
                best_offer1 = offer1
                best_offer2 = offer2

    best_offer1_list.append(best_offer1)
    best_offer2_list.append(best_offer2)
    max_profit_list.append(max_profit)
    
```

```python
print(f'best offer prices are {np.mean(best_offer1_list)} and {np.mean(best_offer2_list)}')

plt.hist(best_offer1_list)
plt.hist(best_offer2_list)
plt.show()
```

### Results
